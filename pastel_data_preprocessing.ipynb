```python
import os
import json
import pandas as pd
from tqdm import tqdm

def load_json_folder_to_df(folder_path):
    all_data = []

    for filename in tqdm(os.listdir(folder_path)):
        if filename.endswith('.json'):
            full_path = os.path.join(folder_path, filename)
            with open(full_path, 'r') as f:
                data = json.load(f)
                data['__source_file'] = filename
                all_data.append(data)

    df = pd.json_normalize(all_data)
    return df
```

```python
folder_path = '/content/drive/My Drive/Colab Notebooks/v2'
```

```python
sentences_train_df = load_json_folder_to_df(folder_path + '/sentences/train/')
sentences_train_df.to_csv('/content/drive/My Drive/DS/sentences_train.csv', index=False)
```

```python
sentences_test_df = load_json_folder_to_df(folder_path + '/sentences/test/')
sentences_test_df.to_csv('/content/drive/My Drive/DS/sentences_test.csv', index=False)
```

```python
sentences_valid_df = load_json_folder_to_df(folder_path + '/sentences/valid/')
sentences_valid_df.to_csv('/content/drive/My Drive/DS/sentences_valid.csv', index=False)
```

```python
stories_train_df = load_json_folder_to_df(folder_path + '/stories/train/')
stories_train_df.to_csv('/content/drive/My Drive/DS/stories_train.csv', index=False)
```

```python
stories_test_df = load_json_folder_to_df(folder_path + '/stories/test/')
stories_test_df.to_csv('/content/drive/My Drive/DS/stories_test.csv', index=False)
```

```python
stories_valid_df = load_json_folder_to_df(folder_path + '/stories/valid/')
stories_valid_df.to_csv('/content/drive/My Drive/DS/stories_valid.csv', index=False)
```

```python
def process_dataframe(df, output_column='output.sentences', persona_column='persona.age'):


    new_df = df[[output_column, persona_column]].copy()
    return new_df
```
